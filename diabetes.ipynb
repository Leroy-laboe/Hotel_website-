# Graphs

# For all Attributes

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('diabetes.csv')

# Loop through all columns except 'Outcome'
features = [col for col in df.columns if col != 'Outcome']

# Create boxplots for each feature vs Outcome
for feature in features:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='Outcome', y=feature, data=df)
    plt.title(f'{feature} by Diabetes Outcome')
    plt.xlabel('Diabetes Outcome (0 = No, 1 = Yes)')
    plt.ylabel(feature)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

# Histogram

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('diabetes.csv')

# Plot histograms for all features
features = [col for col in df.columns if col != 'Outcome']

for feature in features:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=feature, hue='Outcome', kde=True, multiple='stack', palette='Set2')
    plt.title(f'{feature} Distribution by Outcome')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
```

# 2. Pairplot of Key Features

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('diabetes.csv')

# Select key features
selected_features = ['Glucose', 'BMI', 'Age', 'DiabetesPedigreeFunction', 'Outcome']

# Create pairplot
sns.pairplot(df[selected_features], hue='Outcome', corner=True, palette='husl')
plt.suptitle('Pairplot of Key Features Colored by Outcome', y=1.02)
plt.show()
```

# ðŸ“‰ 3. Bar Plot of Average Feature Values by Outcome

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('diabetes.csv')

# Compute average values for each outcome group
mean_values = df.groupby('Outcome').mean().reset_index().melt(id_vars='Outcome')

# Plot barplot
plt.figure(figsize=(12, 6))
sns.barplot(x='variable', y='value', hue='Outcome', data=mean_values)
plt.title('Average Feature Values by Outcome')
plt.xlabel('Feature')
plt.ylabel('Mean Value')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# ðŸ“ˆ 4. Line Graph of Feature Means by Outcome

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('diabetes.csv')

# Calculate mean values
mean_values = df.groupby('Outcome').mean().T

# Plot line graph
plt.figure(figsize=(12, 6))
plt.plot(mean_values[0], marker='o', label='Non-Diabetic (0)', linestyle='--')
plt.plot(mean_values[1], marker='s', label='Diabetic (1)', linestyle='-')
plt.title('Feature Means for Each Outcome Group')
plt.xlabel('Features')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# ðŸ”¥ 5. Correlation Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('diabetes.csv')

# Compute correlation matrix
corr = df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Features')
plt.tight_layout()
plt.show()
```
